# 2. Prepare inversion control file

```python
# Define inversion parameters:

# DISTORTION
DISTORTION = 0 

# TRADE_OFF_PARAM
TRADE_OFF_PARAM = [3]

# Maximum number of iterations
ITERATION = 30

# NUM_THREADS
NUM_THREADS = 6
```

```python
inversion01.write_inversion_control(TRADE_OFF_PARAM = TRADE_OFF_PARAM,
                                   ITERATION = ITERATION,
                                   NUM_THREADS = NUM_THREADS)
```

# 3. Prepare mesh files

```python
outdir = r'meshGen_files'
```

```python
inversion01.analysis_domain
```

```python
mesh01 = femticPy.MeshGen(inversion01.survey,
                         inversion01.analysis_domain, 
                         inversion01.center, 
                         inversion01.mt_coords, 
                         inversion01.mt_data,
                         outdir)
```

mesh01.write_dummy_coast_line()
mesh01.write_dummy_bathy_file()
mesh01.write_dummy_topo_file()

```python
# Step 1: 

mesh01.center = [0.0, 0.0, 0.0]
mesh01.rotation = 0
mesh01.ellipsoids_control = [3,
                     [150, 15.0, 0.0, 0.0, 0.0],
                     [200, 50.0, 0.0, 0.0, 0.0],
                     [2000.0, 500.0, 0.0, 0.0, 0.0]]

mesh01.coast_line = r'.\input_data\coast_line.dat'

mesh01.ellipsoids_observing_sites =  [1, 
                                      10.0, 5]
```

```python
# Step 6: Refine
# this step refines the tetras around the MT sites

mesh01.ellipsoids_obs_sites = [2,                     
                    [25.0, 10.0, 0.6], 
                    [50.0, 30.0, 0.3]]


mesh01.ellipsoids_mtr = [4,
                    [100.0, 25.0, 0.0, 0.7, 0.7],
                    [200, 80.0, 0.0, 0.4, 0.5],
                    [500, 250.0, 0.0, 0.0, 0.3],     
                    [2000.0, 2000.0, 0.0, 0.0, 0.0]]
```

```python
# Step 7: Write mesh to FEMTIC format
# the attr file will determine how tetras will be combined into elements (model parameters)

mesh01.ellipsoids_resistivity_attr = [5,
                    [200.0,  20.0,  0.3,  0.0],
                    [250.0,  20.0,  0.15,  0.0],                  
                    [400.0,  200.0,  0.0,  0.0],   
                    [1000.0,  800.0,  0.0,  0.0], 
                    [2000.0,  2000.0,  0.0,  0.0]] 


mesh01.ellipsoids_resistivity_attr_sites = [3, 
                    [5.0, 1.0],                        
                    [20.0, 2.0],
                    [30.0, 10.0]]
```

```python
mesh01.write_inputs()
```
